```
#1
import pandas as pd
df=pd.read_csv("student-math.csv",sep=";")
df.head(5)
```

```
#2
df["final_grade"]=df['G1'] + df['G2'] + df['G3']
df.head()
```

```
#3
df.drop(['G1','G2','G3'],axis=1,inplace=True)
df.head()
```

```
#4
df.replace({'yes':1,'no':0},inplace=True)
df.head()
```

```
#5
import numpy as np
import matplotlib.pyplot as plt
plt.title("Studytime vs. FinalGrade")
plt.xlabel("Studytime")
plt.ylabel("FinalGrade")
x =df['studytime']
y =df['final_grade']

colors =np.where(x==1,'red',np.where(x==2,'blue',np.where(x==3,'green',np.where(x==4,'violet','red'))))

plt.scatter(x, y, c =colors)
plt.show()
```
